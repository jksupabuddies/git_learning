```python
import pygame
import random as rd
import numpy as np
import math as mth

class Object:
    def __init__(self,mass,*pygameObject):
        self.pyg_s = pygameObject
        self.mass = mass
        self.forces = []
        self.parents = []
        self.children = []
        self.accel = 0
    
    def attachForce(self, force):
        if(isinstance(force,Force) == True):
            if(force not in self.forces):
                self.forces.append(force)
                force.parents.append(self)
            
    def acceleration(self):
        net_sum = np.array([[0],[0]], dtype = "float64")
        for force in self.forces:
            net_sum += force.vectorComp()
        self.accel = net_sum
            
                
                
        

class Force:
    #direction MUST be in radians
    def __init__(self,mag_tude,direction,*name):
        self.direction = direction
        self.mag_tude = mag_tude
        self.name = name
        self.parents = []
        
    def vectorComp(self):
        x = mth.cos(self.direction) * self.mag_tude
        y = mth.sin(self.direction) * self.mag_tude
        vector = np.array([[x],[y]])
        return vector
    



x_max = 1960
y_max = 1080

slowed_time = 0
real_time = 0

currX = x_max/2
currY = y_max

pygame.init()
posX = x_max/2
posY = y_max/2
screen = pygame.display.set_mode((x_max,y_max))
screen.fill("white")
clock = pygame.time.Clock()
running = True
while(running):
    screen.fill("white")
    for event in pygame.event.get():
        key = pygame.key.get_pressed()
        if key[pygame.K_SPACE] or (event.type == pygame.QUIT):
            running = False
    if(key[pygame.K_UP]):
        posY -= 10
    elif(key[pygame.K_DOWN]):
        posY += 10
    elif(key[pygame.K_RIGHT]):
        posX += 10
    elif(key[pygame.K_LEFT]):
        posX -= 10
    cookie = pygame.draw.circle(screen,"black",(posX, posY),10)
    pygame.draw.circle(screen,"blue", (x_max,y_max),10)
    #Time Incrementers
    dst = clock.tick(60)/10000# 60 FPS Limit, essentially assume 60 frames a second
    dt = dst*10 # for acceleration
    pygame.display.update()
pygame.quit()  
```
